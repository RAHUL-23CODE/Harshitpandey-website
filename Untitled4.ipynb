```python
from datasets import load_dataset
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load XSUM
dataset = load_dataset("xsum")

# Load tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("t5-base")
model = T5ForConditionalGeneration.from_pretrained("t5-base")
```

```python
def preprocess(example):
    input_text = "summarize: " + example["document"]
    inputs = tokenizer(input_text, max_length=512, truncation=True, return_tensors="pt")
    labels = tokenizer(example["summary"], max_length=64, truncation=True, return_tensors="pt").input_ids
    return {
        "input_ids": inputs.input_ids[0],
        "attention_mask": inputs.attention_mask[0],
        "labels": labels[0]
    }

tokenized_dataset = dataset["train"].map(preprocess, remove_columns=dataset["train"].column_names)
```

```python
def generate_summary(text):
    input_text = "summarize: " + text
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=64, num_beams=4, early_stopping=True)
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Example
article = dataset["test"][0]["document"]
print("Original Article:", article)
print("Generated Summary:", generate_summary(article))
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import torch

# Load T5-base model (can be vanilla or fine-tuned)
tokenizer = T5Tokenizer.from_pretrained("t5-base")
model = T5ForConditionalGeneration.from_pretrained("t5-base")

# 1. Cosine Similarity Function
def cosine_sim_score(job_description, generated_resume):
    vectorizer = TfidfVectorizer().fit([job_description, generated_resume])
    vectors = vectorizer.transform([job_description, generated_resume])
    score = cosine_similarity(vectors[0:1], vectors[1:2])
    return round(float(score[0][0]), 4)

# 2. Resume Generator using T5
def generate_resume(job_description):
    prompt = "generate resume for: " + job_description
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=150, num_beams=4, early_stopping=True)
    resume = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return resume

# 3. Example Test
job_desc = "We are looking for a data analyst skilled in Python, SQL, Tableau, and ETL."

generated_resume = generate_resume(job_desc)
cosine_score = cosine_sim_score(job_desc, generated_resume)

# 4. Output Result
print("📝 Job Description:\n", job_desc)
print("\n📄 Generated Resume:\n", generated_resume)
print(f"\n📊 Cosine Similarity Score: {cosine_score}")
```

```python
from random import uniform
import matplotlib.pyplot as plt

# Simulated evaluation function
def evaluate_model(model_name, dataset):
    """
    Returns a mock similarity score to simulate evaluation.
    """
    if model_name == "T5-base" and dataset == "XSUM":
        return round(uniform(0.3, 0.45), 3)  # Generic, not tuned
    elif model_name == "T5-base-finetuned-resumes" and dataset == "Resumes":
        return round(uniform(0.75, 0.9), 3)  # Fine-tuned, high quality
    else:
        return round(uniform(0.5, 0.7), 3)

# Compare models
models = ["T5-base", "T5-base-finetuned-resumes"]
datasets = ["XSUM", "Resumes"]
scores = []

for model in models:
    for dataset in datasets:
        score = evaluate_model(model, dataset)
        scores.append({
            "model": model,
            "dataset": dataset,
            "score": score
        })
        print(f"{model} on {dataset}: {score}")

# Visualization
labels = [f"{m['model']} on {m['dataset']}" for m in scores]
values = [m['score'] for m in scores]

plt.figure(figsize=(10, 5))
bars = plt.barh(labels, values, color=["#888", "#1f77b4", "#888", "#2ca02c"])
plt.xlabel("Similarity Score")
plt.title("Model Performance Before and After Fine-Tuning")
plt.xlim(0, 1)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.01, bar.get_y() + bar.get_height()/2,
             f'{width:.3f}', va='center')

plt.tight_layout()
plt.show()
```

```python
from transformers import pipeline
from datasets import load_dataset
import evaluate

# Load evaluation metric
rouge = evaluate.load("rouge")

# Load a small sample from XSUM
dataset = load_dataset("xsum", split="validation[:10]")  # 10 samples for demo

# Load base and fine-tuned summarization pipelines
base_summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base")
finetuned_summarizer = pipeline("summarization", model="mrm8488/t5-base-finetuned-common_gen", tokenizer="mrm8488/t5-base-finetuned-common_gen")

# Run inference
def summarize_and_score(pipeline_model, dataset, label=""):
    preds, refs = [], []
    for item in dataset:
        input_text = item['document']
        reference = item['summary']
        try:
            summary = pipeline_model(input_text, max_length=60, min_length=10, do_sample=False)[0]['summary_text']
            preds.append(summary)
            refs.append(reference)
        except:
            continue
    results = rouge.compute(predictions=preds, references=refs)
    print(f"\n🔍 ROUGE Scores ({label})")
    for k, v in results.items():
        print(f"{k}: {v:.4f}")
    return results

# Evaluate both models
base_scores = summarize_and_score(base_summarizer, dataset, "T5-base")
fine_scores = summarize_and_score(finetuned_summarizer, dataset, "Fine-tuned T5 (CommonGen)")
```

```python
!pip install transformers datasets evaluate torch
!pip install rouge_score
!pip install scikit-learn nltk python-docx pdfplumber
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download the punkt_tab data

from transformers import pipeline
from datasets import load_dataset
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



# Load BLEU, ROUGE, and ATS Scoring Libraries
import evaluate
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")

# 🔹 Load XSUM Dataset (10 samples for speed)
dataset = load_dataset("xsum", split="validation[:10]")

# 🔹 Load Base T5 Model
summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base")

# 🔹 Preprocess Text (Remove Special Characters)
def preprocess_text(text):
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special characters
    return word_tokenize(text.lower())

# 🔹 Compute ATS Score
def compute_ats_score(original_text, generated_summary):
    original_tokens = preprocess_text(original_text)
    summary_tokens = preprocess_text(generated_summary)

    # ✅ Keyword Match Score (how many words overlap)
    matched_keywords = len(set(original_tokens) & set(summary_tokens))
    keyword_score = matched_keywords / len(set(original_tokens)) if original_tokens else 0

    # ✅ TF-IDF Similarity Score
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([original_text, generated_summary])
    tfidf_score = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

    # ✅ Formatting Penalty (checks for non-ATS-friendly content)
    formatting_penalty = 0.1 if any(tag in generated_summary for tag in ["table", "image", "graphic"]) else 0

    # 🔹 Final ATS Score
    ats_score = (0.5 * tfidf_score) + (0.5 * keyword_score) - formatting_penalty
    return round(ats_score * 100, 2)  # Convert to percentage

# 🔹 Evaluate T5-Base Summaries
def evaluate_t5_on_xsum():
    ats_scores = []
    bleu_scores = []
    rouge_scores = []

    for item in dataset:
        original_text = item["document"]
        reference_summary = item["summary"]

        # Generate Summary
        try:
            generated_summary = summarizer(original_text, max_length=60, min_length=10, do_sample=False)[0]["summary_text"]
        except:
            continue

        # Compute ATS Score
        ats = compute_ats_score(original_text, generated_summary)
        ats_scores.append(ats)

        # Compute BLEU Score
        bleu_result = bleu.compute(predictions=[generated_summary], references=[[reference_summary]]) # removed .split() calls
        bleu_scores.append(bleu_result["bleu"])

        # Compute ROUGE Score
        rouge_result = rouge.compute(predictions=[generated_summary], references=[reference_summary])
        rouge_scores.append(rouge_result["rougeL"])

    # 🔹 Print Average Scores
    avg_ats = sum(ats_scores) / len(ats_scores)
    avg_bleu = sum(bleu_scores) / len(bleu_scores)
    avg_rouge = sum(rouge_scores) / len(rouge_scores)

    print(f"\n📊 **Evaluation for T5-Base on XSUM**")
    print(f"✅ Average ATS Score: {avg_ats:.2f}%")
    print(f"✅ Average BLEU Score: {avg_bleu:.4f}")
    print(f"✅ Average ROUGE-L Score: {avg_rouge:.4f}")

# 🚀 Run Evaluation
evaluate_t5_on_xsum()
```
